```python
numeric_df = df.select_dtypes(include=[np.number])

# Run the correlation on the numeric columns
correlation_matrix = numeric_df.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot = True)
plt.rcParams['figure.figsize'] = (10,7)
plt.show()
```
